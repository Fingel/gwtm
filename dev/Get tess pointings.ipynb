```python
import urllib
from urllib.request import urlopen
```

```python
import csv
```

```python
url = "https://tess.mit.edu/wp-content/uploads/Year1-3_pointing_long.csv"
```

```python
import io
csv_response = urllib.request.urlopen(url)
lst = list(csv.reader(io.TextIOWrapper(csv_response)))
```

```python
for row in lst:
    print(row)
```
